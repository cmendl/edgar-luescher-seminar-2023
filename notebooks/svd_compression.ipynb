```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read image from disk
imgc = plt.imread("kitten.jpg")
plt.imshow(imgc);
```

```python
imgc.shape
```

```python
# convert to grayscale and [0, 1] value range
img = np.mean(imgc, axis=2) / 255.0
img.shape
```

```python
plt.imshow(img, cmap="gray");
```

```python
# singular value decomposition
u, s, vh = np.linalg.svd(img)
```

```python
# check
np.allclose(img, (u * s) @ vh)
```

```python
plt.semilogy(s**2 / np.sum(s**2), '.')
plt.ylabel("$\\sigma_j^2$")
plt.xlabel("$j$")
plt.title("normalized singular values");
```

```python
χlist = [len(s), 150, 75, 25, 10]

plt.figure(figsize=(25, 15))
for j, χ in enumerate(χlist):
    plt.subplot(1, len(χlist), j + 1)
    plt.imshow((u[:, :χ] * s[:χ]) @ vh[:χ, :], cmap="gray")
    plt.title("χ = {}".format(χ))
```
